```python
gdp_df = pd.read_csv('../data/gdp_percapita.csv')
gdp_df.head(2)
```

```python
gdp_df.shape
```

```python
gdp_df.info()
```

```python
del gdp_df ['Value Footnotes']
```

```python
gdp_df.columns = ['country', 'year', 'gdp_per_capita']
gdp_df.head(2)
```

```python
gdp_df.year.value_counts()
```

```python
gdp_df.isnull().sum()
```

```python
gdp_df.info()
```

```python
gdp_2020 = gdp_df.loc[gdp_df['year']==2020]
gdp_2020.head(2)  
```

```python
gdp_2020.shape
```

```python
gdp_2020['gdp_per_capita'].hist(bins = 20, color = 'yellow', edgecolor = 'black', figsize = (10,6))
plt.title('2020 gdp per capita - histogram', fontsize = 11);
```

```python
gdp_2020.gdp_per_capita.plot.density(color = 'green')
plt.title ('2020 gdp per capita - density plot')
plt.show()
```

```python
sns.boxplot(data = gdp_2020, x = 'gdp_per_capita')
plt.title ('2020 gdp per capita - boxplot')
plt.show()
```

```python
sns.violinplot(x=gdp_2020['gdp_per_capita'])
plt.title ('2020 gdp per capita - violin plot')
plt.show()
```

```python
gdp_2020.gdp_per_capita.median()
```

```python
gdp_1990 = gdp_df.loc[gdp_df['year']==1990]
gdp_1990.head(2)
```

```python
gdp_2000 = gdp_df.loc[gdp_df['year']==2000]
gdp_2020.head(2) 
```

```python
gdp_2010 = gdp_df.loc[gdp_df['year']==2010]
gdp_2020.head(2) 
```

```python
gdp_decades = pd.concat([gdp_1990, gdp_2000, gdp_2010, gdp_2020])
gdp_decades.head(2)
```

```python
sns.boxplot(data = gdp_decades, y = 'gdp_per_capita', x = 'year')
plt.title ('1990 - 2020 - boxplot')
plt.show()
```

```python
sns.barplot(data = gdp_decades, y = 'gdp_per_capita', x = 'year') 
plt.title ('1990 - 2020 - sns.barplot')
plt.show()
```

```python
gdp_decades['year'].value_counts().sort_values(ascending = True).plot(kind = 'bar', color = 'orange', edgecolor = 'black');
```

```python
gdp_decades.plot.scatter(x = 'year', y = 'gdp_per_capita')
plt.show()
```

```python
sns.regplot(x = gdp_decades['year'], y = gdp_decades['gdp_per_capita'])
plt.show()
```

```python
gdp_100k = gdp_decades.loc[gdp_decades['gdp_per_capita'] > 100000] 
gdp_100k
```
